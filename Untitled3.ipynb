```python
import matplotlib.pyplot as plt
import pandas as pd

data = [['a',10],['b',12],['c',15],['d',11],['e',14],['f',16]]



df =  pd.DataFrame(data, columns=['nome','idade'])

```

```python
plt.boxplot(df['idade']);
```

```python
df.describe()
```

```python
plt.figure(figsize=(10,5))
plt.pie(df['idade'], autopct='%.2f%%',explode=[0.1,0,0,0,0,0,]);
plt.legend(df['nome'])
plt.show()
```
