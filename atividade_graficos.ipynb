```python
import matplotlib.pyplot as plt

quantidade = (40,65,80,55)

meses = ['jan-2023','fev-2023','março-2023','abril-2023']

fig, ax = plt.subplots(figsize=(10,5))

ax.bar(meses,quantidade)


ax.set_xlabel('meses')
ax.set_ylabel('valor gasto')
ax.set_title('valor historico dos graficos')
ax.legend(title='gasto com combustivel');
```
